```
import bs4
import re #re= regular expression
import nltk # a pre-developed library for data analysis techniques START PRE_PROCESSING
import urllib.request
```

```
text_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/COVID-19_vaccine').read()
```

```
print(text_data)
```

```
obj2 = bs4.BeautifulSoup(text_data,"lxml") #parsing the url document, creating the object for beautiful soup 
```

```
print(obj2)
```

```
text = " "
for paragraph in obj2.find_all('p'):
  text += paragraph.text
```

```
print(text)
```

```
text = re.sub(r'\[[0-9]*\]',' ',text)
```

```
print(text)
```

```
cleaner_text = text.lower()
cleaner_text = re.sub(r'\d',' ',cleaner_text)#removes decimal digits
cleaner_text = re.sub(r'\W',' ',cleaner_text)#short form of re.sub [a-zA-Z0-9]
cleaner_text = re.sub(r'\s+',' ',cleaner_text) #remove extra spaces
```

```
cleaner_text
```

```
nltk.download("punkt") 
```

```
sentences = nltk.sent_tokenize(text)
```

```
sentences
```

```
nltk.download("stopwords")
```

```
stopwords = nltk.corpus.stopwords.words('english')
```

```
wordstoCount = {} #dictionary not list
for word in nltk.word_tokenize(cleaner_text):
  if word not in stopwords:
    if word not in wordstoCount.keys():
      wordstoCount[word] = 1
    else:
        wordstoCount[word]+= 1
```

```
print(wordstoCount)
```

```
maximum_count = max(wordstoCount.values())
```

```
print(maximum_count) #converting counts to weights
```

```
for key in wordstoCount.keys(): #converting counts to weights
  wordstoCount[key] = wordstoCount[key]/maximum_count
```

```
print(wordstoCount[key])
```

```
sentscore = {} #taking wordto count score to sentence tokens
for sentence in sentences:
  for word in nltk.word_tokenize(sentence.lower()):
    if word in wordstoCount.keys():
      if len(sentence.split(" "))<25:
        if sentence not in sentscore.keys():
          sentscore[sentence] = wordstoCount[word]
        else:
           sentscore[sentence] += wordstoCount[word] 
```

```
import heapq
best_sentences = heapq.nlargest(5,sentscore,key=sentscore.get)
```

```
print("--------------------------------------------------")
for sentence in best_sentences:
  print(sentence)
```
